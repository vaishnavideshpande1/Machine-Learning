1. Generate a numpy array with 2 columns, where first column contains numbers from 0 to 5 and second column is [0, 1, 4, 9, 16, 25]

2. Check the shape of your array. It should be 6x2.

3. Put the array into a data frame, with column labels ‘x’ and ‘y’ 

4. Plot ‘x’ versus ‘y’ 

5. Get average of elements in second column of the data frame


```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
MyMatrix=np.array([[0,0],[1,1],[2,4],[3,9],[4,16],[5,25]])
print('The shapeof the array is ',MyMatrix.shape)
colLabels=['x','y']
MyDF=pd.DataFrame(MyMatrix,columns=colLabels)
plt.plot(MyDF.x,MyDF.y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('MyMatrix')
Average_y=sum(MyDF.y)/len(MyDF.y)
print('the average of elements in second column is ',Average_y)
```
